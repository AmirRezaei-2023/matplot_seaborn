شاید از خود بپرسید دلیل وجود سیمکالون در انتهای خطوط کد بالا چیست؟ مگر قرار نبود پایتون سیمکالون نداشته باشد؟ حالا بریم سراغ پاسخ.

در جوپیتر، هنگامی که شما اسم متغیر را در سلول می‌نوشتید، مقدار متغیر در خروجی نمایش داده می‌شد. یعنی اگر متغیر ما لیست پایتون، آرایه نامپای، رشته یا انواع دیگر متغیر باشد، با نوشتن اسم آن در سلول و اجرای آن، مقدار آن متغیر در خروجی نمایش داده می‌شود.

این ویژگی خوبی است که ما هم تا اینجا بارها از آن استفاده کرده‌ایم. اما اگر موقعیتی پیش بیاید که ما نیاز به دیدن مقدار متغیر نداشته باشیم، چه باید کرد؟ در این حالات با قرار دادن سیمکالون در انتهای آن خط، مقدار آن متغیر در خروجی نمایش داده نمی‌شود.

برای نمایش نمودارها در این فصل، ما شی شکل (در ادامه با آن آشنا خواهید شد) را در انتهای سلول می‌نویسیم و سلول را اجرا می‌کنیم. اگر سیمکالن در انتهای خط نگذاریم، علاوه بر عکس نمودار، محتوای شیء شکل، در خروجی نمایش داده می‌شود. چاپ مقدار شیء شکل ضرری برای کسی ندارد و ما برای زیبا‌تر شدن خروجی‌ها، آن را نمایش نمی‌دهیم

عناصر شکل و صفحه محورها، دو عنصر اساسی طرح‌ها در مت‌پلات هستند. شکل (figure) سطح بالاترین عنصریست که همهٔ عناصر دیگر یک طرح (plot) را در بر می‌گیرد. به این ترتیب وقتی می‌خواهیم نموداری رسم کنیم، اولین چیزی که ساخته می‌شود یک شکل است که عناصر دیگر داخل آن قرار می‌گیرند. در کد بالا، شکل به صورت غیر مستقیم ساخته شده است و می‌توان آن را به ترتیب زیر هم ساخت.

```python
fig = plt.figure()
```

عنصر دیگری که در رسم کردن نمودارها نقش کلیدی دارد، صفحه محور است. هر نموداری که رسم می‌کنیم داخل یک صفحه محور قرار می‌گیرد. صفحه محور عناصری مانند محورها، صفحه مختصات، عنوان نمودار و سایر جزئیات آن را نشان می‌دهد. در شکل بالا، یک صفحه محور داشتیم که نمودار را در همان رسم کردیم. اما می‌توان شکل‌هایی شامل چند صفحه محور هم ساخت. برای این منظور می‌توانیم یک صفحه محور را در یک مختصات خاص از شکل بسازیم و یا شکل را به جدولی از صفحه‌محورها تقسیم کنیم. در ادامه به هر دو روش خواهیم پرداخت. پیش از آن لازم است کمی درباره دو نوع شیوه کار با مت‌پلات بدانیم.

از آنجایی که مت‌پلات در ابتدا به عنوان جایگزین پایتونی متلب نوشته شد، بسیاری از دستورات آن از نظر نحوی مشابه متلب تعریف شده‌اند. اصلی‌ترین رابط برای استفاده از این دستورات متلبی رابط pyplot است. شیوه کار این رابط به شکل وضعیت‌مند است. به این ترتیب که همواره شکل و صفحه محور اخیر را نگه می‌دارد و متدی از رابط را که صدا می‌شود روی آن اجرا می‌کند. در کدی که در ابتدا نوشتیم یک شکل ابتدایی و یک صفحه محور داخل آن به صورت غیر‌مستقیم تعریف شده‌اند و با صدا زدن متد plot ،نمودارهای مربوطه روی آن رسم می‌شوند.

با صدا زدن متد plt.axes می‌توانیم یک صفحه محور جدید بسازیم و مختصات نقطهٔ پایین چپ و طول و عرض آن را نسبت به اندازهٔ شکل در قالب یک لیست ورودی دهیم. پس از تعریف این صفحه محور، به عنوان صفحه محور اخیر در نظر گرفته می‌شود و تا زمانی که با صفحه محور دیگری جایگزین نشود، متدهای مختص به یک صفحه محور رابط روی آن اعمال می‌شوند. در کد زیر، یک نمونه استفاده از آن را مشاهده می‌کنید.

```python
plt.figure()
plt.axes()
plt.plot(x, np.exp(x))
plt.show()
```

```python
plt.figure()
plt.axes()
plt.plot(x, np.exp(x))
plt.axes([0.15, 0.3, 0.5, 0.3])
plt.plot(x, np.log(x), color='blue', linestyle='dashed')
```

همان‌طور که پیش‌تر گفتیم، صفحه محورها را می‌توان به شکل یک جدول هم داخل شکل قرار داد. برای این منظور می‌توانیم از متد plt.subplot استفاده کنیم. به این متد تعداد سطرها و ستون‌های جدولِ صفحه محورها و شماره صفحه محور داخل جدول را می‌دهیم تا ساخته شود (شماره‌ها به صورت ستونی و سپس سطری از بالا چپ جدول شروع می‌شوند و مشابه متلب شماره نخستین صفحه محور برابر یک است). توجه داشته باشید که هر صفحه محور پس از ساخت به عنوان صفحه محور اخیر در نظر گرفته می‌شود. کد زیر، نمونه استفاده از این روش را نشان می‌دهد

```python
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x))
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), linestyle='dashdot', color='red')
plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x), color='green', linestyle='dashed')
plt.subplot(2, 2, 4)
plt.plot(x, x**2, color='magenta')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.15, 0.3, 0.5, 0.3])
ax1.plot(x, np.exp(x))
ax2.plot(x, np.log(x), color='blue', linestyle='dashed')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), color='red', linestyle='dashdot')
ax3.plot(x, np.exp(x), color='green', linestyle='dashed')
ax4.plot(x, x**2, color='magenta', linestyle='solid')
```

برای این منظور، راه ساده‌تری هم وجود دارد و آن استفاده از تابع subplots در رابط pyplot است. این تابع یک شکل و یک آرایه از صفحه محورهای داخل آن با ابعادی که ورودی گرفته، می‌سازد. به این ترتیب می‌توانیم با استفاده از آرایه به صفحه محورها دسترسی داشته باشیم.

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
ax[0,2].plot(x, x**2, color='magenta', linestyle='solid')
```
